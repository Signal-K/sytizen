```
import lightkurve as lk
import matplotlib.pyplot as plt
```

## Kepler-186

```
TIC = 'Kepler-186'
available_data_select = lk.search_lightcurve(TIC)[0:3]
lc_collection = available_data_select.download_all().stitch()
%matplotlib inline
fig, ax = plt.subplots(figsize = (8, 4))
lc_collection.plot(ax = ax, linewidth = 0, marker = 0, color = 'green', markersize = 1, alpha = 0.8)
```

```
%matplotlib inline
bin_time = 15/24/60
lc_collection_binned = lc_collection.bin(bin_time)
lc_collection_binned.plot(color = 'gold', lw = 0, marker = '.')
```

```
period = 10
t0 = 260
```

```
lc_phased = lc_collection.fold(period = period, epoch_time = t0)
%matplotlib inline
fig, ax = plt.subplots(figsize = (8, 4))
lc_phased.plot(ax = ax, linewidth = 0, marker = 0, color = 'pink', markersize = 1, alpha = 0.8)
```
