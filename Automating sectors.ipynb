```python
TIC = 'Kepler-69c'
available_data_all = lk.search_lightcurve(TIC, author="SPOC")
available_data_all
sector_data = lk.search_lightcurve(TIC, author = 'SPOC', sector = 74)
lc = sector_data.download()
lc.plot(linewidth = 0, marker = '.', color = 'midnightblue', alpha = 0.5)

lc_stitched = lc.normalize()
lc_stitched.plot(color = 'gold', lw = 0, marker = '.')

bin_time = 15/24/60
lc_collection_binned = lc_stitched.bin(bin_time)

# Modify the figure and axes properties for transparency
fig, ax = plt.subplots()
ax.set_facecolor('none')
fig.patch.set_alpha(0)

# Plotting the binned lightcurve with transparent background
lc_collection_binned.plot(ax=ax, color = 'red', lw = 0, marker = '.')

plt.show()
```
