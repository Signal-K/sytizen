```python
import matplotlib.pyplot as plt
import lightkurve as lk
```

```python
TIC = 'TIC 55525572'
available_data_select = lk.search_lightcurve(TIC, author = 'SPOC')[1:4]
lc_collection = available_data_select.download_all().stitch()
```

```python
lc_collection.plot()
```

Binning -> Downsizing the data samples to get rid of noise (takes the mean of data points)
<br />
This makes it easier to spot the transit events in the graph

```python
bin_time = 15/24/60
lc_collection_binned = lc_collection.bin(bin_time)
```

```python
lc_collection_binned.plot()
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
lc_collection.plot(ax = ax, linewidth = 0, marker = 'o', color = 'gold', markersize = 1)
lc_collection_binned.plot(ax = ax, linewidth = 0, marker = '*', color = 'red', markersize = 2)
```
